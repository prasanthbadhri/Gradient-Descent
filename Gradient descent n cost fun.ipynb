## gradient descent

```python
# its used to find the best line...
# minimising the cost function...(cost fun or loss fun  should be nearer to zero)
```

### gradient descent is used to reduce the cost fun by taking the best parameters...

### learning rate : L = 0.001,0.0001.....(steps to minimization) (baby steps)

### iterations : n = 500 or 1000 or 2000 etc (no of iterations)

### derivative of m nd b : (derivatives are used to minimize the error)

### upgrade m nd b :

#### finally, we get the very less change in our cost m and b. 
#### so, we will figure out our cost..
#### And we will take those m and b to our prediction best fit line

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
def gradient_descent(x,y,learning_rate = 0.001,iterations = 200):
    # random m and b
    m_curr = 1
    b_curr = 1
    
    n = len(x)
    
    for i in range(iterations):
        y_pred = m_curr * x + b_curr
        cost = 1/n * sum([val**2 for val in y-y_pred])   #  MSE --> mean square error...    
        
        # derivative of m and b
        
        md = -(2/n)*sum(x*(y-y_pred))
        bd = -(2/n)*sum(y-y_pred)

        # update m nd b
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
        print("iterations: ",i,"--- m= ",m_curr," b= ",b_curr," cost= ",cost)
        
```

```python
x = np.array([1,2,3,4,5,6,7])
y= np.array([12,20,34,41,52,63,72])

gradient_descent(x,y)
```

```python
x = np.array([1,2,3,4,5,6,7])
y= np.array([12,20,34,41,52,63,72])

gradient_descent(x,y,learning_rate=0.01) #like this we can change the learning rate to obtain the best fit line...
```
